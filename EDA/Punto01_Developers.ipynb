```python
import pandas as pd
```

```python
games_path = '../ETL/Data_Extracted/DataFrame_games.parquet'
GamesDF = pd.read_parquet(games_path)
```

<p>Extraigo del dataframe unicamente las columnas que sirven para responder lo que se pide en este caso <b>'developer','release_date','genres'</b></p>

```python
#Filtro  por solo las columnas que  voy a usar
GamesDF = GamesDF[['developer','release_date','genres']]

```

```python
GamesDF.dropna()
```

```python
# Genero una columna para saber si tiene la opcion de jugar gratis
# Aqui aplico una funcion lambda a cada fila que devuelve 1 si contiene el elemento free to play 0 si no
GamesDF['free_to_play'] = GamesDF['genres'].apply(lambda x: 1 if 'Free to Play' in x else 0)
```

```python
# Asi va quedando
GamesDF
```

```python
#Elimino la columna Genre porque ya no la necesito mas
GamesDF = GamesDF.drop(columns=['genres'])
```

```python
# Convierto la serie 'release_date' a cadena (string) porque voy a tratar la fecha como string
GamesDF['release_date'] = GamesDF['release_date'].astype(str)

# Definir una expresión regular para extraer el año
pattern = r'\b(\d{4})\b'
# Este patron busca 4 numeros concecutivos que serian el año que quiero extraer

# Aplicar la expresión regular y extraer el año
GamesDF['year'] = GamesDF['release_date'].str.extract(pattern)
```

```python
# quedaron 2 registros que no se pudieron extraer el año
GamesDF['year'].isnull().sum()
```

```python
#Borror la fecha de lanzamiento porque ya tengo el año que necesito
GamesDF = GamesDF.drop(columns=['release_date'])
#Borrar registros nulos, las fechas que quedaron nulas no me sirven para filtrar
GamesDF = GamesDF.dropna()
```

```python
# Implemento la funcion que hace el filtro por desarrollador
def developer(desarrollador: str):
    # Filtrar el DataFrame por el desarrollador específico
    df_desarrollador = GamesDF[GamesDF['developer'] == desarrollador]

    # Agrupar por año y contar la cantidad de items y el contenido free
    info_por_año = df_desarrollador.groupby('year').agg(
        cantidad_items=('developer', 'count'),
        contenido_free=('free_to_play', lambda x: f"{(sum(x) / len(x) * 100):.2f}%")
    ).reset_index()

    return info_por_año
```

```python
# prueba
resultado = developer('playstation')
resultado
```
