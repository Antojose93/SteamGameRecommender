#### **UsersRecommend( año : int )**

Devuelve el top 3 de juegos MÁS recomendados por usuarios para el año dado. (reviews.recommend = True y comentarios positivos/neutrales)

Ejemplo de retorno: [{"Puesto 1" : X}, {"Puesto 2" : Y},{"Puesto 3" : Z}]

```python
import pandas as pd
```

```python
games_path = '../ETL/Data_Extracted/DataFrame_games.parquet'
GamesDF = pd.read_parquet(games_path)
```

```python
GamesDF = GamesDF.rename(columns={'id': 'item_id'})
GamesDF
```

```python
Reviews_path = '../ETL/Data_Extracted/DataFrame_reviews.parquet'
ReviewsDF = pd.read_parquet(Reviews_path)
```

```python
ReviewsDF['recommend'] = ReviewsDF['recommend'].replace({True: 1, False: 0})
ReviewsDF
```

```python
# Convierto la serie 'release_date' a cadena (string) porque voy a tratar la fecha como string
GamesDF['release_date'] = GamesDF['release_date'].astype(str)

# Definir una expresión regular para extraer el año
pattern = r'\b(\d{4})\b'
# Este patron busca 4 numeros concecutivos que serian el año que quiero extraer

# Aplicar la expresión regular y extraer el año
GamesDF['year'] = GamesDF['release_date'].str.extract(pattern)
```

```python
#Se extraen las columnas para generar un nuevo dataframe y correr la función
GamesDF= GamesDF[["item_id","app_name","year",'developer']]
ReviewsDF= ReviewsDF[["item_id","recommend"]]
BestDevDF= ReviewsDF.merge(GamesDF, on="item_id")
BestDevDF= BestDevDF.drop("item_id", axis= 1)
BestDevDF= BestDevDF.drop_duplicates()
BestDevDF.reset_index(drop=True, inplace=True)
BestDevDF
```

```python
BestDevDF['year'] = pd.to_numeric(BestDevDF['year'], errors='coerce')
```

```python
BestDevDF['year'] = str(BestDevDF['year'])
```

```python
BestDevDF.info()
```

```python
def best_developer_year(year):
    # Filtrar el DataFrame por el año dado
    year_df = BestDevDF[BestDevDF['year'] == year]

    # Ordenar el DataFrame por la cantidad de recomendaciones en orden descendente
    sorted_df = year_df.sort_values(by='recommend', ascending=False)

    # Seleccionar los primeros tres juegos de la lista ordenada
    top_3_games = sorted_df.head(3)

    return top_3_games['developer'].values
```

```python
print(best_developer_year(2013))
```
